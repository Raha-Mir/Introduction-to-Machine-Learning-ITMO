### Train the linear regression model to predict the values and enter the specified parameters.

#### 0. Import necessary libraries

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

#### 1. Create the DataFrame manually

```python
X = np.array([16, 9, 12, 21, 8, 23, 25, 11, 3, 18])  # Number of people in line
Y = np.array([29, 27, 34, 43, 25, 45, 62, 29, 7, 29])  # Time spent in store (min)
```

#### 2.

> Find the sample mean $\bar{X}$:

image.png

```python
x = np.mean(X)
print(f'the sample mean X: {x}')
```


> Find the sample mean $\bar{Y}$:

image.png

```python
y = np.mean(Y)
print(f'the sample mean Y: {y}')
```


> Calculate the coefficient $\theta_0$

> Calculate the coefficient $\theta_1$

image.png

```python
theta1 = sum([(X[i]-x)*(Y[i]-y) for i in range(len(X))])/sum([(X[i]-x)**2 for i in range(len(X))])
theta0 = y - theta1*x

print (f'theta0: {round(theta0,2)}')
print (f'theta1: {round(theta1,2)}')
```


> Estimate the model accuracy by calculating R-squared statistics, $R^2$:

image.png

```python
R_squre = 1 - sum([(Y[i]-theta0-theta1*X[i])**2 for i in range(len(X))]) / sum([(Y[i]-y)**2 for i in range(len(Y))])
print(f'R^2: {round(R_squre, 2)}')
```
