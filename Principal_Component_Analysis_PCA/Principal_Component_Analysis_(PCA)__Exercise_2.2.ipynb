#### 0. Import necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import minmax_scale
```

#### 1. Load the Data

```python
score_matrix = pd.read_csv('https://storage.yandexcloud.net/lms-itmo-ru-files-27a87tyf/machine_learning/task2/task2.2/data/_reduced/X_reduced_473.csv', encoding = 'utf-8', delimiter=';', header=None)
score_matrix.head() # Display the first 5 rows of the result
```

```python
loading_matrix = pd.read_csv('https://storage.yandexcloud.net/lms-itmo-ru-files-27a87tyf/machine_learning/task2/task2.2/data/_loadings/X_loadings_473.csv', encoding = 'utf-8', delimiter=';', header=None)
loading_matrix.head() # Display the first 5 rows of the result
```

#### 2. Matrix multiplication

```python
# Reconstruct the original data using PCA formula
reconstructed = score_matrix @ loading_matrix.T

plt.matshow(reconstructed)
plt.title("Reconstructed Party Logo")
plt.show()
```
